```python
import pandas as pd
from sqlalchemy import create_engine
```

# CSV File cleanup

```python
# Import files
csvfile = "DataFiles/USVideos.csv"

us_videos = pd.read_csv(csvfile, parse_dates=['trending_date', 'publish_time'])
us_videos.head()
```

```python
us_videos.columns
```

```python
# Get rid of unnecessary columns
us_videos_df = us_videos[['trending_date', 'title', 'channel_title', 'category_id',
       'publish_time', 'views', 'likes', 'dislikes', 'comment_count']].copy()
us_videos_df.head()
```

```python
# Data Transformation: Rename Columns
us_videos_df = us_videos_df.rename(columns={"trending_date":"Trending_Date", "title":"Title", "channel_title":"Channel_Title",
                            "category_id":"Category_ID", "publish_time":"Publish_Time", "views":"View_Count", 
                             "likes":"Like_Count", "dislikes":"Dislike_Count", "comment_count":"Comment_Count"})
```

```python
# Data Transformation: Change to datetime
us_videos_df['Trending_Date']= pd.to_datetime(us_videos_df["Trending_Date"])
```
